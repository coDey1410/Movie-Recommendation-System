```python
import numpy as np
import ast
import pandas as pd
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert(movies.iloc[0].genres) #this shows the first movie genres in a list form
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert2(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.iloc[0]['cast']
```

```python
def fetch_director(obj):
    L=[]
    
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())    #convert the overview string into a list
```

```python
movies['overview']
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags'][0]        # tags of the first movie
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))   #convert the list to string
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
#__________vectorization______________
# we will vectorize the words and the the distance between the words ans check its similarity
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()   # the 5000 most frequent words removing the stop words
```

```python
len(cv.get_feature_names() )
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
ps.stem('lovely')  # an example of stemming
```

```python
new_df['tags']=new_df['tags'].apply(stem)   # stemming done.
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()  # re transform after stemming
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors) 


#similarity matrix is an array 
#of cosine similairty between a movie and all other movies inclusing itself
```

```python
similarity.shape
```

```python
list(enumerate(similarity[0]))  # we used enumerate to store the corresponding index and the cosine value then we convert them into a list
```

```python
 sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[0:11] 
    
    # thats how we sort it without scrambling the indexes
    # top 10 similar movies using slicing [0:11]
```

```python
similarity[0]  #similarity vector for the first movie
```

```python
def recommend(movie):
    
    movie_index= new_df[new_df['title']==movie].index[0]  #finds the index of the movie
    
     #now checks the similarity of the movie and stores the similarity vector in the distances vector 
    distances=similarity[movie_index] 
    
    movie_list= sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[0:11] 
   
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
